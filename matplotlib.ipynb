```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,3,8,5,5,10]
plt.plot(x,y)
plt.xlabel('mehvare x')
plt.ylabel('mehvare y')
plt.title('my graph')
plt.show()
```

```python
x1 = [1,2,3,4,5,6]
y1 = [1,3,8,5,5,10]
x2 = [5,2,3,2,5,4]
y2 = [1,6,8,5,2,10]
plt.plot(x1,y1,label="xy1")
plt.plot(x2,y2,label="xy2",color="r")
plt.bar(x1,y1,color="g")
plt.legend()
plt.xlabel('mehvare x')
plt.ylabel('mehvare y')
plt.title('my graph')
plt.show()
```

```python
x1 = [1,2,3,4,5,6]
y1 = [1,3,8,5,5,10]
x2 = [5,2,3,2,5,4]
y2 = [1,6,8,5,2,10]
plt.scatter(x1,y1,label="xy1")
plt.scatter(x2,y2,label="xy2",color="r")
plt.legend()
plt.xlabel('mehvare x')
plt.ylabel('mehvare y')
plt.title('my graph')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y1 = [1,3,8,5,5,10]
y2 = [1,6,8,5,2,10]
plt.stackplot(x,y1,y2)

plt.xlabel('mehvare x')
plt.ylabel('mehvare y')
plt.title('my graph')
plt.show()
```

```python
slices = [2,3,6,8,3]
ls = ['x1','x2','x3','x4','x5']
plt.pie(slices,labels=ls,shadow=True,explode=(0,0.1,0,0,0,))
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation as FA
from random import randint
```

```python
count = 0
x_values = []
y_values = []

def animate(i):
    x_values.append(count)
    y_values.append(randint(0,10))
    count += 1
    # plt.cal()
    plt.plot(x_values,y_values)

# get current fiqure ==> gcf
FA(plt.gcf(),animate,interval=1000)

plt.tight_layout()
plt.show()
```
