```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## load image as gray

```python
img = cv2.imread('input.jpg',0)
img.ndim
```

```python
cv2.imshow('img',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('input.jpg')
```

## convert image to gray with cv2

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.ndim
```

```python
gray_image.shape
```

```python
cv2.imshow('img',gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

## convert image to gray with numpy

```python
im_gray = img[:,:,0]//3 + img[:,:,1]//3 + img[:,:,2]//3
```

```python
cv2.imshow('img',im_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

## show gray image with matplotlib

```python
plt.imshow(im_gray,cmap="gray")
```

## make subplots for show many image in one figure

```python
b,g,r = cv2.split(img)
f,(row1,row2) =plt.subplots(2,2)

row1[0].imshow(img[:,:,::-1])
row1[0].set_title('img')

row1[1].imshow(b,cmap="gray")
row1[1].set_title('b')

row2[0].imshow(g,cmap="gray")
row2[0].set_title('g')

row2[1].imshow(r,cmap="gray")
row2[1].set_title('r')
```

## split & merge image channels

```python
b,g,r = cv2.split(img)
img = cv2.merge([b,g,r])
plt.imshow(img)
```

## cropping

```python
crop_image = img[250:450,250:550]
plt.imshow(crop_image[:,:,::-1])
```

```python
h,w,channels = img.shape
```

```python
new_image = img[h//3:3*h//4,w//3:3*w//4]
plt.imshow(new_image[:,:,::-1])
```

## image save

```python
cv2.imwrite('new_image.png',new_image)
```

```python
img = cv2.imread('input.jpg')
b,g,r = cv2.split(img)
new_r = cv2.add(r,90)
new_b = cv2.add(b,90)
img = cv2.merge([new_b,g,new_r])
plt.imshow(img[:,:,::-1])
```

```python
img = cv2.imread('input.jpg')
b,g,r = cv2.split(img)

f,(row1,row2) =plt.subplots(2,2)

row1[0].imshow(img[:,:,::-1])
row1[0].set_title('img')


new_r = cv2.add(r,90)
red_img = cv2.merge([b,g,new_r])
row1[1].imshow(red_img)
row1[1].set_title('b')

new_g = cv2.add(g,90)
green_img = cv2.merge([b,new_g,r])
row2[0].imshow(green_img)
row2[0].set_title('g')

new_b = cv2.add(b,90)
blue_img = cv2.merge([new_b,g,r])
row2[1].imshow(blue_img)
row2[1].set_title('r')

```
