```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('checkerboard_18x18.png')
plt.imshow(img)
img
```

## Load the two images

```python
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')
```

```python
plt.imshow(image1[:,:,::-1])
```

```python
plt.imshow(image2[:,:,::-1])
```

## Resize the images to a common size

```python
image1 = cv2.resize(image1, (350, 300))
image2 = cv2.resize(image2, (350, 300))
```

```python
plt.imshow(image1[:,:,::-1])
```

```python
plt.imshow(image2[:,:,::-1])
```

## Create frames (black backgrounds) for each image

```python
frame1 = np.zeros((400, 400, 3), dtype='uint8')
frame2 = np.zeros((400, 400, 3), dtype=np.uint8)
```

```python
plt.imshow(frame1)
```

## Place the images within the frames

```python
frame1[50:350, 0:350] = image1
frame2[50:350, 50:400] = image2
```

## Concatenate the two frames horizontally to create the combined image

```python
combined_image_h = cv2.hconcat([frame1, frame2])
combined_image_v = cv2.vconcat([frame1, frame2])
```

## Save the combined image

```python
cv2.imwrite('combined_image_h.jpg', combined_image_h)
cv2.imwrite('combined_image_v.jpg', combined_image_v)
```

## Display the combined image

```python
plt.imshow(combined_image_h[:,:,::-1])
```

```python
plt.imshow(combined_image_v[:,:,::-1])
```
