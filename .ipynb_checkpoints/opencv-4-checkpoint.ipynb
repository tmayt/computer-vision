```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input.jpg')

img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imshow('img',img_hsv)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
h = img_hsv[:,:,0]
s = img_hsv[:,:,1]
v = img_hsv[:,:,2]

f,(row1,row2) =plt.subplots(2,2)

row1[0].imshow(img[:,:,::-1])
row1[0].set_title('img')

row1[1].imshow(h,cmap="gray")
row1[1].set_title('hue')

row2[0].imshow(s,cmap="gray")
row2[0].set_title('saturation')

row2[1].imshow(v,cmap="gray")
row2[1].set_title('value')
```

```python
new_s = cv2.add(s,100)
new_hsv_img = cv2.merge([h, new_s, v])
new_rgb_img = cv2.cvtColor(new_hsv_img,cv2.COLOR_HSV2RGB)
plt.imshow(new_rgb_img)
```

```python
plt.imshow(img[:,:,::-1])
```

```python
new_v = cv2.add(v,100)
new_hsv_img = cv2.merge([h, s, new_v])
new_rgb_img = cv2.cvtColor(new_hsv_img,cv2.COLOR_HSV2RGB)
plt.imshow(new_rgb_img)
```
